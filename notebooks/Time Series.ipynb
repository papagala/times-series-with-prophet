```python
# Python
import pandas as pd
from fbprophet import Prophet
from fbprophet.plot import plot_plotly, plot_components_plotly

%matplotlib inline
```

# Crear función reutilizable para graficar

```python
def plot_time_series(file,forecast_time=30,seasonality_mode='additive',freq='D',changepoint_prior_scale=0.05):
    df = pd.read_csv(file)
    m = Prophet(seasonality_mode=seasonality_mode,changepoint_prior_scale=changepoint_prior_scale)
    m.fit(df)
    future = m.make_future_dataframe(periods=forecast_time,freq=freq)
    forecast = m.predict(future)
    forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
    fig1 = m.plot(forecast)
    fig2 = m.plot_components(forecast)
    fig3 = plot_plotly(m, forecast)
    fig4 = plot_components_plotly(m, forecast)
    return fig1, fig2, fig3, fig4

```

# 1 Peyton Manning

Serie de tiempo de el número de vistas de la página de Wikipedia de Peyton Manning. Originalmente se utilizó el paquete Wikipediatrend de R. Esto provee un ejemplo muy claro de las características de prophet como múltiple estacionalidad, tasas cambiantes, y la habilidad del modelo para los días especiales (es decir los días de playoff o superbowl).

```python
results=plot_time_series(file='../data/examples/example_wp_log_peyton_manning.csv',forecast_time=365)
```

# 2

Por defecto, la librería prophet ajusta estacionalidades aditivas, lo que quiere decir que los efectos de la estacionalidad son agregados a la tendencia para obtener la predicción. La serie de tiempo del número de pasajeros aereos es un ejemplo cuando la estacionalidad aditiva no funciona. 

```python
results=plot_time_series('../data/examples/example_air_passengers.csv',forecast_time=60,freq='MS')
```

La serie de tiempo tiene claramente un ciclo anual, pero la estacionalidad en esta predicción es muy grande en el comienzo de la serie de tiempo y muy pequeña al final. En esta serie de tiempo, la estacionalidad no es un factor constante aditivo tal como lo asume prohet, en lugar de ello crece con la tendencia. Esta es una estacionalidad multiplicativa.

```python
results=plot_time_series('../data/examples/example_air_passengers.csv',forecast_time=60,seasonality_mode='multiplicative',freq='MS')
```

Con esto, los efectos de vacaciones también serán multiplicativos.
